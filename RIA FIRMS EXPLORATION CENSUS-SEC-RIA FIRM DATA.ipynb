```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}

# Loop through the cities_pd and run a lat/long search for each city
for index, row in CO_IRA_FIRMS_DF.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"

    place = row['Business Name']
    

    # update address key value
    params['input'] = f"{place}"
    params['inputtype'] = 'textquery'
    params['fields'] = 'formatted_address,geometry'  
    # make request
    co_ria_info = requests.get(base_url, params=params)
    
    # print the cities_lat_lng url, avoid doing for public github repos in order to avoid exposing key
    print(co_ria_info.url)
    
    # convert to json
    co_ria_info = co_ria_info.json()

    try:
        CO_IRA_FIRMS_DF.loc[index, "Address"] = co_ria_info["candidates"][0]["formatted_address"]
        CO_IRA_FIRMS_DF.loc[index, "Lat"] = co_ria_info["candidates"][0]["geometry"]["location"]["lat"]
        CO_IRA_FIRMS_DF.loc[index, "Lng"] = co_ria_info["candidates"][0]["geometry"]["location"]["lng"] 
    
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
# Visualize to confirm lat lng appear
#CO_IRA_FIRMS_DF.head()
```
